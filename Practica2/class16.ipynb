```python
df_limpio-isna().sum()
NA_val = df_limpio.isna().sum()
def filtro_na(na, umbral=0.3): 
    col_pass=[]
    for i in na.key():
        if na[i]/df_limpio.shape[0]<umbral:
            col_pass-append(i)
    return col_pass
df_limpio = df_limpio(filtro_na(NA_val))
```
